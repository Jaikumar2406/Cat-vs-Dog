```python
import tensorflow as tf
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tensorflow.keras.utils import load_img, img_to_array
```

```python
trainin = tf.keras.utils.image_dataset_from_directory(
    "animals/",
    batch_size=16,
    image_size=(128,128),
    shuffle=True,
    seed=123,
    subset="training",
    validation_split=0.2)
```

```python
validation = tf.keras.utils.image_dataset_from_directory(
    "animals/",
    batch_size=16,
    image_size=(128,128),
    shuffle=True,
    seed=123,
    subset="validation",
    validation_split=0.2)
```

```python
train_ds = trainin.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
validation = validation.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense , Dropout , GlobalAveragePooling2D
from tensorflow.keras.models import Model
```

```python
base_model = MobileNetV2(input_shape=(128,128)+(3,),
                         include_top=False,
                         weights='imagenet',)
```

```python
base_model.trainable = False
```

```python
x = base_model.output
x = GlobalAveragePooling2D()(x)
x=  Dropout(0.3)(x)
output= Dense(1, activation = "sigmoid")(x)
```

```python
model =Model(inputs = base_model.input ,outputs = output)
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'],
)
```

```python
model.summary()
```

```python
history_fine = model.fit(
    train_ds,
    validation_data=validation,
    epochs=10
)
```

```python
model.evaluate(validation)
```

```python
img_path = 'shutterstock_218749363.jpg'
img = load_img(img_path, target_size=(128,128))
img_array = img_to_array(img) / 255.0
img_array = tf.expand_dims(img_array, 0) 
plt.imshow(img)

pred = model.predict(img_array)[0][0]
print("Prediction:", "Dog" if pred > 0.5 else "Cat", "| Confidence:", round(pred, 2))
```
